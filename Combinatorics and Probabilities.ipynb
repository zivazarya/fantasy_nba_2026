```python
import math
import numpy as np
```

```python
def win_exactly(n,k,p):
    if k<0 or k>n:
        return 0
    return (math.comb(n, k)*(p**(k))*((1-p)**(n-k)))
```

```python
#Probability to win up to k times out of n with probability of p to win each time
def win_at_most(n,k,p):
    if k>=n:
        return 0
    sum=0
    for i in range (k+1):
        sum=sum+win_exactly(n,i,p)
    return sum
```

```python
#Probability to win at least k times out of n with probability of p to win each time
def win_at_least(n,k,p):
    if k<=0:
        return 1
    if k>n:
        return 0
    return 1-win_at_most(n,k-1,p)
```

```python
#Probability to win at least k times out of n, with probability p1 to win in n1 tries and probability p2 to win in n2 tries
def win_at_least_2(n,k,n1,p1,p2):
    sum=0
    for i in range (n1+1):
        if i+n-n1>=k:
            sum=sum+win_exactly(n1,i,p1)*win_at_least(n-n1,k-i,p2)
    return sum    
```

```python
def win_at_least_k_out_of_n(n,k,nums_percs):
    nums_percs[:,1].astype(int)
    prob_sum=0
    
    def win_at_least_rec_init(n,k,nums_percs):
        num=int(nums_percs[0,0])
        perc=nums_percs[0,1]
        nums_percs_next = np.delete(nums_percs, (0), axis=0)
        n_next=n-num
        for i in range (num+1):
            p_next=win_exactly(num,i,perc)
            win_at_least_rec(n,k-i,nums_percs_next,p_next)
            
                
    def win_at_least_rec(n,k,nums_percs,p):
        nonlocal prob_sum
        if nums_percs.shape[0]==1:
            num=int(nums_percs[0,0])
            perc=nums_percs[0,1]
            p_final=win_at_least(num,k,perc)
            prob_sum=prob_sum+p*p_final
        else:
            num=int(nums_percs[0,0])
            perc=nums_percs[0,1]
            nums_percs_next=nums_percs
            nums_percs_next=np.delete(nums_percs_next, (0), axis=0)
            for i in range (num+1):
                p_next=p*win_exactly(num,i,perc)
                win_at_least_rec(n,k-i,nums_percs_next,p_next)

    win_at_least_rec_init(n,k,nums_percs)
    return prob_sum
```
