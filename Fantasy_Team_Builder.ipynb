```python
# Based on Medium
games_per_week = 82/22
benchmarks_50 = {"p/g": 581,
                 "r/g": 217,
                 "a/g": 130,
                 "s/g": 36.5,
                 "b/g": 24,
                 "to/g": 58,
                 "3/g": 70,
                 "fg%": 0.491462,
                 "ft%": 0.811981}
benchmarks_75 = {"p/g": 642.25,
                 "r/g": 234.75,
                 "a/g": 143,
                 "s/g": 41,
                 "b/g": 28.25,
                 "to/g": 52,
                 "3/g": 70,
                 "fg%": 0.505771,
                 "ft%": 0.836771}
```

```python
benchmark_1 = pd.Series(benchmarks_50)
benchmark_2 = pd.Series(benchmarks_75)
```

```python
col_names = ["Player", "Team", "Pos", "p/g", "3/g", "r/g", "a/g", "s/g", "b/g", "fg%", "fga/g", "ft%", "fta/g", "to/g"]
```

```python
cumm_cat = list(benchmarks_50.keys())
perc_cat = ["fg%", "fga/g", "ft%", "fta/g"]
```

```python
proj_df = pd.read_excel('projection_12_10_25.xlsx')
```

```python
proj_df = proj_df[col_names]
proj_df["Player"] = proj_df["Player"].str.lower()
proj_df
```

## Team Builder

#### initialization

```python
curr_team = pd.DataFrame(columns=col_names)
```

```python
def add_player(team, player):
  p = proj_df[proj_df["Player"] == player]
  i = p.index
  for col in col_names[3:]:
    if "%" not in col:
      #p[col] = p[col] * games_per_week
      p.loc[i, col] = p.loc[i, col] * games_per_week
  p.loc[i, "fgm/g"] = p.loc[i, "fg%"] * p.loc[i, "fga/g"]
  p.loc[i, "ftm/g"] = p.loc[i, "ft%"] * p.loc[i, "fta/g"]
  return pd.concat([team, p])
```

```python
def calc_week(team):
  week_stats = team[cumm_cat].sum()
  fg_sum = team[["fga/g", "fgm/g"]].sum()
  week_stats["fg%"] = fg_sum["fgm/g"] / fg_sum["fga/g"]
  ft_sum = team[["fta/g", "ftm/g"]].sum()
  week_stats["ft%"] = ft_sum["ftm/g"] / ft_sum["fta/g"]
  return week_stats
```

## Theoretical strategies analysis

```python
from scipy.stats import binom
```

```python
def k_or_more(k, n):
  return 1 - binom.cdf(k - 1, n, 0.5)
```

```python
# 25% FT, 50% for rest
0.25 * k_or_more(4, 8) + 0.75 * k_or_more(5, 8)
```

```python
# 25% FT, 75% Reb, 50% rest
0.25 * 0.75 * k_or_more(3, 7) + 0.25 * 0.25 * k_or_more(4, 7) + \
0.75 * 0.75 * k_or_more(4, 7) + 0.75 * 0.25 * k_or_more(5, 7)
```

```python
# 25% FT, 65% Reb+BLK, 50% rest
0.25 * binom.pmf(2, 2, 0.65) * k_or_more(2, 6) + 0.25 * binom.pmf(1, 2, 0.65) * k_or_more(3, 6) + \
0.25 * binom.pmf(0, 2, 0.65) * k_or_more(4, 6) + 0.75 * binom.pmf(2, 2, 0.65) * k_or_more(3, 6) + \
0.75 * binom.pmf(1, 2, 0.65) * k_or_more(4, 6) + 0.75 * binom.pmf(0, 2, 0.65) * k_or_more(5, 6)
```

```python
# 5% FT, 70% Reb+BLK, 50% rest
0.05 * binom.pmf(2, 2, 0.70) * k_or_more(2, 6) + 0.05 * binom.pmf(1, 2, 0.70) * k_or_more(3, 6) + \
0.05 * binom.pmf(0, 2, 0.70) * k_or_more(4, 6) + 0.95 * binom.pmf(2, 2, 0.70) * k_or_more(3, 6) + \
0.95 * binom.pmf(1, 2, 0.70) * k_or_more(4, 6) + 0.95 * binom.pmf(0, 2, 0.70) * k_or_more(5, 6)
```

```python
# 5% FT, 60% PTS+Reb+BLK+FG%, 50% rest
0.05 * binom.pmf(4, 4, 0.60) * k_or_more(0, 4) + 0.05 * binom.pmf(3, 4, 0.60) * k_or_more(1, 4) + \
0.05 * binom.pmf(2, 4, 0.60) * k_or_more(2, 4) + 0.05 * binom.pmf(1, 4, 0.60) * k_or_more(3, 4) + \
0.05 * binom.pmf(0, 4, 0.60) * k_or_more(4, 4) + 0.95 * binom.pmf(4, 4, 0.60) * k_or_more(1, 4) + \
0.95 * binom.pmf(3, 4, 0.60) * k_or_more(2, 4) + 0.95 * binom.pmf(2, 4, 0.60) * k_or_more(3, 4) + \
0.95 * binom.pmf(1, 4, 0.60) * k_or_more(4, 4)
```
